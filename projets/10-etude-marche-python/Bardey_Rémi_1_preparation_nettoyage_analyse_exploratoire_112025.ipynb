# Calcul des distances France > Pays

```python
la_poule_qui_chante.info()
```

```python
# Installation des bibliothèques et modules pour le calcul des distances

!pip install -q geopy countryinfo deep-translator

from deep_translator import GoogleTranslator
from countryinfo import CountryInfo
from geopy.distance import geodesic
```

```python
# traduire les pays Fr en En

translator = GoogleTranslator(source='fr', target='en')
_cache = {}

def tr_fr_en(name: str) -> str:
    if name in _cache:
        return _cache[name]
    en = translator.translate(str(name).strip())
    _cache[name] = en
    return en

la_poule_qui_chante['pays_en'] = la_poule_qui_chante['pays'].apply(tr_fr_en)
```

```python
# Affiche toutes les lignes et toutes les colonnes - Vérification des traductions

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

print(la_poule_qui_chante[['pays', 'pays_en']])
```

```python
# Formatage de l'affichage (pas au complet)

pd.reset_option('display.max_rows')
pd.reset_option('display.max_columns')
```

```python
# Calcul des distances depuis la France

FR_LAT, FR_LON = 46.2276, 2.2137  # France (centroïde)

# Définir lat et lon pour chaque pays
def get_latlon(country_en: str):
    try:
        info = CountryInfo(str(country_en)).info()
        ll = info.get("latlng")
        if isinstance(ll, (list, tuple)) and len(ll) == 2:
            return float(ll[0]), float(ll[1])
    except Exception:
        pass
    return np.nan, np.nan

# Récupérer lat/lon à partir de la colonne traduite 'pays_en'
la_poule_qui_chante[["lat", "lon"]] = la_poule_qui_chante["pays_en"].apply(
    lambda x: pd.Series(get_latlon(x))
)

# Calculer la distance géodésique en km depuis la France
def dist_from_france(lat, lon):
    if pd.isna(lat) or pd.isna(lon):
        return np.nan
    return geodesic((FR_LAT, FR_LON), (lat, lon)).km

la_poule_qui_chante["distance_km"] = la_poule_qui_chante.apply(
    lambda r: dist_from_france(r["lat"], r["lon"]), axis=1
)

# Contrôle rapide
print("Lignes sans distance (coordonnées manquantes) :",
      la_poule_qui_chante["distance_km"].isna().sum())

la_poule_qui_chante[["pays", "pays_en", "distance_km"]].head(10)
```

```python
la_poule_qui_chante.info()
```

```python
# Voir quels pays n'ont pas de coordonnées
la_poule_qui_chante.loc[la_poule_qui_chante['distance_km'].isna(), ['pays', 'pays_en']].sort_values('pays')
```

```python
# Corrections ciblées des pays EN + recalcul pour les lignes manquantes

FIX = {
    "Benign": "Benin",
    "China - Hong Kong SAR": "Hong Kong",
    "China - Macau SAR": "Macao",
    "Congo": "Republic of the Congo",
    "Dominic": "Dominica",
    "Eswatini": "Swaziland",
    "Grenade": "Grenada",
    "Iran (Islamic Republic of)": "Iran",
    "Netherlands (Kingdom of)": "Netherlands",
    "United Kingdom of Great Britain and Northern Ireland": "United Kingdom",
    "Democratic Republic of Congo": "Democratic Republic of the Congo",
    "Swiss": "Switzerland",
    "Suede": "Sweden",
    "Czechia": "Czech Republic",
    "Timor-Leste": "East Timor",
    "Micronesia (Federated States of)": "Micronesia",
    "Gambia": "Gambia",
    "North Macedonia": "North Macedonia",
    "Montenegro": "Montenegro",
    "Sao Tome and Principe": "Sao Tome and Principe",
    "Myanmar": "Myanmar",
}

# Remplacer uniquement sur les lignes qui n'ont pas de distance
mask_na = la_poule_qui_chante['distance_km'].isna()
la_poule_qui_chante.loc[mask_na, 'pays_en'] = (
    la_poule_qui_chante.loc[mask_na, 'pays_en'].replace(FIX)
)

# Recalculer lat/lon pour ces lignes seulement
def get_latlon(country_en: str):
    try:
        info = CountryInfo(str(country_en)).info()
        ll = info.get("latlng")
        if isinstance(ll, (list, tuple)) and len(ll) == 2:
            return float(ll[0]), float(ll[1])
    except Exception:
        pass
    return np.nan, np.nan

idx = la_poule_qui_chante.index[mask_na]
new_coords = la_poule_qui_chante.loc[idx, 'pays_en'].apply(get_latlon).apply(pd.Series)
new_coords.columns = ['lat', 'lon']
la_poule_qui_chante.loc[idx, ['lat', 'lon']] = new_coords

# Recalculer la distance pour ces mêmes lignes
FR_LAT, FR_LON = 46.2276, 2.2137

def dist_from_france(lat, lon):
    if pd.isna(lat) or pd.isna(lon):
        return np.nan
    return geodesic((FR_LAT, FR_LON), (lat, lon)).km

la_poule_qui_chante.loc[idx, 'distance_km'] = la_poule_qui_chante.loc[idx].apply(
    lambda r: dist_from_france(r['lat'], r['lon']), axis=1
)

# Bilan - test
print("Encore manquants :", la_poule_qui_chante['distance_km'].isna().sum())
la_poule_qui_chante.loc[la_poule_qui_chante['distance_km'].isna(), ['pays','pays_en']].sort_values('pays').head(20)
```

```python
# Supprimer les lignes sans distance
la_poule_qui_chante = la_poule_qui_chante.dropna(subset=['distance_km']).reset_index(drop=True)

print("Lignes restantes :", len(la_poule_qui_chante))
```

```python
# Suppression colonnes inutiles

la_poule_qui_chante.drop(columns=['pays_en', 'lat', 'lon'], inplace=True)
```
